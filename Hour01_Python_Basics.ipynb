````python
def f(x):
    """
    Divide a number by two.

    Parameters
    ----------

    x : number
        That which is to be halved.

    Returns
    -------
    float
        Input argument divided by 2.
    """
    return x / 2

for i in range(5):
    print(i, f(i))
````
